##Gráficos

```python
def graph(x, y, col):
  fig, ax = plt.subplots()
  ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
  ax.set_title(col)
  ax.set_ylabel("Distribuição")
  ax.set_xlabel("Valores")
  fig.show()
  
```

```python
def getXY(colum):
  x = []
  y = []
  for i in range(len(colum)):
    y.append(colum.iloc[i])
    x.append(colum.index[i])
  return x,y
```

```python
for col in data.columns:
  temp = data[col].value_counts()
  x,y = getXY(temp)
  graph(x,y, col)
```
